# Particle Counts

```python
counts_SSoG = counts_AM_S1# + counts_AM_SP + counts_AM_S2
#
counts_rest_SoG = counts_AM_N1 + counts_AM_N2 + counts_AM_N3 + counts_AM_C1
#
avg_SSoG, avg_rest_SoG = np.mean(counts_SSoG), np.mean(counts_rest_SoG)
sum_SSoG, sum_rest_SoG = counts_SSoG.sum().iloc[0], counts_rest_SoG.sum().iloc[0]
print('Avg. counts SSoG = ', avg_SSoG)
print('Avg. counts rest of the SoG = ', avg_rest_SoG)
#
print('Total Sum counts SSoG = ', sum_SSoG)
print('Total Sum counts rest of the SoG = ', sum_rest_SoG)

#
print('Avg. ratio between SSoG and the rest of the SoG basin is = ', avg_SSoG/avg_rest_SoG)
print('ratio between sum of SSoG and the rest of the SoG basin is = ', sum_SSoG/sum_rest_SoG)
```
