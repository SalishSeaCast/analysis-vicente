# Method for Avg. Concentration at S4-1.5 and Ratio between S4-1.5 and NG 

```python
# May 2017 (S4-1.5)
may_2017_S4_15_depth = np.array([2, 25, 50, 75, 150, 300])
may_2017_S4_15 = {
    "28": np.array([4.85, 3.15, 7.35, 12.1, 12.7, 5.79]),
    "47": np.array([123, 79.7, 176, 336, 256, 133]),
    "99": np.array([29.5, 20.2, 41.9, 75.5, 41.7, 32.2]),
    "153": np.array([0.67, 0.48, 2.20, 1.49, 0.69, 0.69]),
    "183": np.array([0.15, 0.15, 0.17, 0.16, 0.17, 0.17]),
    "209": np.array([25.7, 25.4, 28.8, 27.7, 130, 28.6])
}

# April 2018 (S4-1.5)
apr_2018_S4_15_depth = np.array([2, 24, 59, 73, 160, 307])
apr_2018_S4_15 = {
    "28": np.array([6.14, 3.04, 4.85, 2.58, 3.90, 5.34]),
    "47": np.array([261, 111, 182, 80.4, 134, 176]),
    "99": np.array([90.0, 35.8, 46.9, 29.0, 42.3, 54.7]),
    "153": np.array([3.76, 1.06, 1.42, 1.14, 1.12, 1.75]),
    "183": np.array([0.16, 0.15, 0.15, 0.15, 0.15, 0.15]),
    "209": np.array([26.6, 26.2, 26.2, 87.1, 26.0, 26.0])
}

# August 2018 (S4-1.5)
aug_2018_S4_15_depth = np.array([4, 26, 54, 70, 147, 294])
aug_2018_S4_15 = {
    "28": np.array([5.86, 7.55, 7.36, 3.03, 5.17, 5.97]),
    "47": np.array([133, 228, 211, 82.3, 180, 244]),
    "99": np.array([50.7, 83.5, 85.5, 26.2, 69.1, 100]),
    "153": np.array([2.33, 2.47, 2.94, 0.77, 2.02, 2.49]),
    "183": np.array([0.16, 0.16, 0.16, 0.16, 0.23, 0.16]),
    "209": np.array([91.38, 26.63, 90.02, 26.83, 39.94, 27.00])
}

# August 2018 (NG)
aug_2018_NG_depth = np.array([2, 25, 50, 75, 150, 330])
aug_2018_NG = {
    "28": np.array([0.64, 0.66, 0.76, 0.51, 0.48, 0.74]),
    "47": np.array([7.7, 7.7, 7.8, 7.6, 7.9, 7.5]),
    "99": np.array([1.49, 1.99, 2.02, 1.96, 2.04, 1.56]),
    "153": np.array([0.12, 0.16, 0.16, 0.16, 0.16, 0.13]),
    "183": np.array([0.12, 0.17, 0.17, 0.16, 0.17, 0.13]),
    "209": np.array([21.1, 28.3, 28.7, 27.8, 29.0, 22.2])
}

# --- DATA FILTERING MASKS ---
mask_may = (may_2017_S4_15_depth >= 50) & (may_2017_S4_15_depth <= 300)
mask_apr = (apr_2018_S4_15_depth >= 59) & (apr_2018_S4_15_depth <= 307)
mask_aug_s4 = (aug_2018_S4_15_depth >= 54) & (aug_2018_S4_15_depth <= 294)
mask_aug_ng = (aug_2018_NG_depth >= 50) & (aug_2018_NG_depth <= 330)

# The updated list of congeners
congeners = ["28", "47", "99", "153", "183", "209"]

# --- CALCULATIONS & OUTPUT ---

print("1. Baseline S4-1.5 Average Concentrations (pg/kg)")
print("-" * 65)
print(f"{'Congener':<10} | {'May 2017 (50-300m)':<25} | {'Apr 2018 (59-307m)':<25}")
print("-" * 65)

for congener in congeners:
    may_mean = np.mean(may_2017_S4_15[congener][mask_may])
    apr_mean = np.mean(apr_2018_S4_15[congener][mask_apr])
    print(f"PBDE {congener:<5} | {may_mean:<25.2f} | {apr_mean:<25.2f}")

print("\n")
print("2. August 2018 Spatial Ratios (S4-1.5 / NG)")
print("-" * 65)
print(f"{'Congener':<10} | {'Ratio (S4/NG) ± SE':<25}")
print("-" * 65)

for congener in congeners:
    # NG Stats
    ng_vals = aug_2018_NG[congener][mask_aug_ng]
    mu_ng = np.mean(ng_vals)
    se_ng = np.std(ng_vals, ddof=1) / np.sqrt(len(ng_vals))
    
    # S4 Stats
    s4_vals = aug_2018_S4_15[congener][mask_aug_s4]
    mu_s4 = np.mean(s4_vals)
    se_s4 = np.std(s4_vals, ddof=1) / np.sqrt(len(s4_vals))
    
    # Ratio Calculation
    ratio = mu_s4 / mu_ng
    se_ratio = ratio * np.sqrt((se_s4 / mu_s4)**2 + (se_ng / mu_ng)**2)
    
    print(f"PBDE {congener:<5} | {ratio:.2f} ± {se_ratio:.2f}")
```

```python
def finder2(target_lat, target_lon,mask=mask):
    lat = mask.nav_lat
    lon = mask.nav_lon
    distance = ((lat - target_lat)**2 + (lon - target_lon)**2)**0.5
    nearest_index = distance.argmin()
    return int(nearest_index // lon.shape[1]), int(nearest_index % lon.shape[1])
```

```python
from scipy.optimize import curve_fit
#
sum_s415 = np.array([381919+552334, 846268+656320, 405950+386180, 236621+248517, 174034+182518])
volume = 109.14092697168105 * 10**9 # m^3
time = 61 * 12 # 61 days and 12 outputs per day
np_per_day = 96 # particles/day
nyears = 5
years_arr = np.arange(2018, 2013, -1)

# Exponential saturation model
def curve(t, goal, decay, to):
    return goal * (1. - np.exp(-(t+to)/decay))

# Dictionaries to iterate through dataframes
scenarios = {
    'Minimum': min_yearly,
    'Best': best_yearly,
    'Maximum': max_yearly
}
congeners = ['BDE-28+33', 'BDE-47', 'BDE-99', 'BDE-153', 'BDE-183', 'BDE-209']
target_concentrations = []
gpp_list = []
#
for congener in congeners:
    for scenario_name, df in scenarios.items():
        
        sumit = np.zeros(nyears)
        contrib = np.zeros(nyears)
        
        # Calculate annual and cumulative contributions
        for iy, year in enumerate(years_arr):
            discharge_val = df.loc[year, congener]
            gpp = discharge_val / np_per_day
            contrib[iy] = gpp * (sum_s415[iy] / volume / time)
            #
            if year == 2018:
                gpp_2018 = gpp            
            #
            if iy == 0:
                sumit[iy] = contrib[iy]
            else:
                sumit[iy] = sumit[iy-1] + contrib[iy]
        #        
        gpp_list.append({
            'Congener': congener,
            'Scenario': scenario_name,
            'GPP_2018': gpp_2018
        })        
        # 3. Curve Fitting for Equilibrium and Decay
        guess_c = sumit[-1] * 1.5 
        p0_guess = [guess_c, 4, 1.3] 
        
        try:
            # Fitting points from Year 2 to Year 5
            popt, pcov = curve_fit(curve, np.arange(nyears-1), sumit[1:], p0=p0_guess, maxfev=10000)
            c_inf = popt[0]
            tau = popt[1]
        except RuntimeError:
            c_inf = np.nan
            tau = np.nan
            
        target_concentrations.append({
            'Congener': congener,
            'Scenario': scenario_name,
            'C_inf': c_inf,
            'Tau': tau
        })

#
results_df = pd.DataFrame(target_concentrations)
gpp_df = pd.DataFrame(gpp_list)

# 1. Format and Pivot GPP Data
gpp_df['GPP_Formatted'] = gpp_df.apply(lambda row: f"{row['GPP_2018']:.3e}", axis=1)
pivot_gpp = gpp_df.pivot(index='Congener', columns='Scenario', values='GPP_Formatted')
pivot_gpp = pivot_gpp[['Minimum', 'Best', 'Maximum']] 

# 2. Format and Pivot Concentration Data
results_df['Display Format'] = results_df.apply(
    lambda row: f"{row['C_inf']:.2e} (τ ~ {row['Tau']:.1f} yrs)" 
    if pd.notnull(row['C_inf']) else "Fit Failed", 
    axis=1
)
pivot_results = results_df.pivot(index='Congener', columns='Scenario', values='Display Format')
pivot_results = pivot_results[['Minimum', 'Best', 'Maximum']] 

print("Equivalent Grams Per Particle (gpp) for 2018 [g/particle]")
print("-" * 75)
print(pivot_gpp)
print("\n")
print("Estimated Asymptotic Concentrations [g m^-3] & Effective Decay Timescales [S1]")
print("-" * 75)
print(pivot_results)
```

```python
def polygon_definition_data_colloidal(ds, region_polygon,  depth_min=50):
#
    lons = ds['lon'].values
    lats = ds['lat'].values
    depths = ds['z'].values
    statuses = ds['status'].values
    #
    points_flat = np.column_stack((lons.ravel(), lats.ravel()))
    poly_path = Path(region_polygon)
    # points inside polygon
    inside_mask_flat = poly_path.contains_points(points_flat)
    inside_mask = inside_mask_flat.reshape(lons.shape)
    # >= 50 mask
    deep_mask = depths >= depth_min
    #deep_mask = depths >= depth_min
    # Colloidal status 
    status_mask = np.isin(statuses, 2) 
    #
    valid_mask = inside_mask & deep_mask & status_mask
    #
    counts = np.sum(valid_mask, axis=0) 
    #
    times = ds['time'].values
    time_index = np.nanmax(times, axis=0) 
    #
    return pd.DataFrame({'count': counts}, index=time_index)
```

```python
def surface_area_region(polygon, mesh_path='/ocean/vvalenzuela/MOAD/grid2/mesh_mask202108_TDV.nc', mask_path='/ocean/vvalenzuela/MOAD/grid2/mesh_mask202108_TDV.nc'):
    mesh = xr.open_dataset(mesh_path)
    mask_ds = xr.open_dataset(mask_path)
    cell_area = mesh['e1t'][0] * mesh['e2t'][0] 
    surface_mask = mask_ds['tmask'][0, 0]
    
    x = cell_area['x']  
    y = cell_area['y'] 
    
    xx, yy = np.meshgrid(x, y)
    
    polygon_mask_2d = np.array([
        [polygon.contains(Point(xx[j, i], yy[j, i])) for i in range(len(x))]
        for j in range(len(y))
    ])
    
    combined_mask_2d = polygon_mask_2d & (surface_mask.values == 1)
    mask_da = xr.DataArray(combined_mask_2d, dims=cell_area.dims, coords=cell_area.coords)
    surface_area_in_polygon = cell_area.where(mask_da).sum().item()
    
    return surface_area_in_polygon
```
