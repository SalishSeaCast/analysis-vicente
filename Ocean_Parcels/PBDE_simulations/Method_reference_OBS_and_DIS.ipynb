```python
def finder2(target_lat, target_lon,mask=mask):
    lat = mask.nav_lat
    lon = mask.nav_lon
    distance = ((lat - target_lat)**2 + (lon - target_lon)**2)**0.5
    nearest_index = distance.argmin()
    return int(nearest_index // lon.shape[1]), int(nearest_index % lon.shape[1])
```

```python
from scipy.optimize import curve_fit
#
sum_s415 = np.array([381919+552334, 846268+656320, 405950+386180, 236621+248517, 174034+182518])
volume = 109.14092697168105 * 10**9 # m^3
time = 61 * 12 # 61 days and 12 outputs per day
np_per_day = 96 # particles/day
nyears = 5
years_arr = np.arange(2018, 2013, -1)

# Exponential saturation model
def curve(t, goal, decay, to):
    return goal * (1. - np.exp(-(t+to)/decay))

# Dictionaries to iterate through dataframes
scenarios = {
    'Minimum': min_yearly,
    'Best': best_yearly,
    'Maximum': max_yearly
}
congeners = ['BDE-28+33', 'BDE-47', 'BDE-99', 'BDE-153', 'BDE-183', 'BDE-209']
target_concentrations = []
gpp_list = []
#
for congener in congeners:
    for scenario_name, df in scenarios.items():
        
        sumit = np.zeros(nyears)
        contrib = np.zeros(nyears)
        
        # Calculate annual and cumulative contributions
        for iy, year in enumerate(years_arr):
            discharge_val = df.loc[year, congener]
            gpp = discharge_val / np_per_day
            contrib[iy] = gpp * (sum_s415[iy] / volume / time)
            #
            if year == 2018:
                gpp_2018 = gpp            
            #
            if iy == 0:
                sumit[iy] = contrib[iy]
            else:
                sumit[iy] = sumit[iy-1] + contrib[iy]
        #        
        gpp_list.append({
            'Congener': congener,
            'Scenario': scenario_name,
            'GPP_2018': gpp_2018
        })        
        # 3. Curve Fitting for Equilibrium and Decay
        guess_c = sumit[-1] * 1.5 
        p0_guess = [guess_c, 4, 1.3] 
        
        try:
            # Fitting points from Year 2 to Year 5
            popt, pcov = curve_fit(curve, np.arange(nyears-1), sumit[1:], p0=p0_guess, maxfev=10000)
            c_inf = popt[0]
            tau = popt[1]
        except RuntimeError:
            c_inf = np.nan
            tau = np.nan
            
        target_concentrations.append({
            'Congener': congener,
            'Scenario': scenario_name,
            'C_inf': c_inf,
            'Tau': tau
        })

#
results_df = pd.DataFrame(target_concentrations)
gpp_df = pd.DataFrame(gpp_list)

# 1. Format and Pivot GPP Data
gpp_df['GPP_Formatted'] = gpp_df.apply(lambda row: f"{row['GPP_2018']:.3e}", axis=1)
pivot_gpp = gpp_df.pivot(index='Congener', columns='Scenario', values='GPP_Formatted')
pivot_gpp = pivot_gpp[['Minimum', 'Best', 'Maximum']] 

# 2. Format and Pivot Concentration Data
results_df['Display Format'] = results_df.apply(
    lambda row: f"{row['C_inf']:.2e} (τ ~ {row['Tau']:.1f} yrs)" 
    if pd.notnull(row['C_inf']) else "Fit Failed", 
    axis=1
)
pivot_results = results_df.pivot(index='Congener', columns='Scenario', values='Display Format')
pivot_results = pivot_results[['Minimum', 'Best', 'Maximum']] 

print("Equivalent Grams Per Particle (gpp) for 2018 [g/particle]")
print("-" * 75)
print(pivot_gpp)
print("\n")
print("Estimated Asymptotic Concentrations [g m^-3] & Effective Decay Timescales [S1]")
print("-" * 75)
print(pivot_results)
```

```python
def polygon_definition_data_colloidal(ds, region_polygon,  depth_min=50):
#
    lons = ds['lon'].values
    lats = ds['lat'].values
    depths = ds['z'].values
    statuses = ds['status'].values
    #
    points_flat = np.column_stack((lons.ravel(), lats.ravel()))
    poly_path = Path(region_polygon)
    # points inside polygon
    inside_mask_flat = poly_path.contains_points(points_flat)
    inside_mask = inside_mask_flat.reshape(lons.shape)
    # >= 50 mask
    deep_mask = depths >= depth_min
    #deep_mask = depths >= depth_min
    # Colloidal status 
    status_mask = np.isin(statuses, 2) 
    #
    valid_mask = inside_mask & deep_mask & status_mask
    #
    counts = np.sum(valid_mask, axis=0) 
    #
    times = ds['time'].values
    time_index = np.nanmax(times, axis=0) 
    #
    return pd.DataFrame({'count': counts}, index=time_index)
```

```python
def surface_area_region(polygon, mesh_path='/ocean/vvalenzuela/MOAD/grid2/mesh_mask202108_TDV.nc', mask_path='/ocean/vvalenzuela/MOAD/grid2/mesh_mask202108_TDV.nc'):
    mesh = xr.open_dataset(mesh_path)
    mask_ds = xr.open_dataset(mask_path)
    cell_area = mesh['e1t'][0] * mesh['e2t'][0] 
    surface_mask = mask_ds['tmask'][0, 0]
    
    x = cell_area['x']  
    y = cell_area['y'] 
    
    xx, yy = np.meshgrid(x, y)
    
    polygon_mask_2d = np.array([
        [polygon.contains(Point(xx[j, i], yy[j, i])) for i in range(len(x))]
        for j in range(len(y))
    ])
    
    combined_mask_2d = polygon_mask_2d & (surface_mask.values == 1)
    mask_da = xr.DataArray(combined_mask_2d, dims=cell_area.dims, coords=cell_area.coords)
    surface_area_in_polygon = cell_area.where(mask_da).sum().item()
    
    return surface_area_in_polygon
```
