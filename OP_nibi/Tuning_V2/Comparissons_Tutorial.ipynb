# Functions for Volume and Colloidal particle counts

Observations for both NG and S4-15 stations have more consistent data from 50 m depth, with more samples in time. Therefore, calculations and comparissons will be done from 50 m to the bottom.

Volume

```python
def volume_region_50m_bottom(polygon):
    mesh = xr.open_dataset(path['mask'])
    volume = mesh['volume']
    mask = xr.open_dataset(path['mask'])['tmask'][0]
    #
    depths = mesh['gdept_1d'][0] 
    depth_condition = depths >= 50
    # 
    depth_mask_3d = depth_condition.values[:, np.newaxis, np.newaxis]
    #    
    x = volume['x']  
    y = volume['y'] 
    #
    xx, yy = np.meshgrid(x, y)
    #
    # 2D Polygon Mask
    polygon_mask_2d = np.array([
        [polygon.contains(Point(xx[j, i], yy[j, i])) for i in range(len(x))]
        for j in range(len(y))
    ])

    # to 3D
    nz = volume.sizes['z']
    polygon_mask_3d = np.repeat(polygon_mask_2d[np.newaxis, :, :], nz, axis=0)

    # inside polygon  + in water + > 50 m
    combined_mask = (polygon_mask_3d & (mask.values == 1) & depth_mask_3d)

    # Applying mask
    mask_da = xr.DataArray(combined_mask, dims=volume.dims, coords=volume.coords)
    volume_in_polygon_water = volume.where(mask_da).sum().item()
    
    return volume_in_polygon_water
```

```python
def volume_region_50m_150m(polygon):
    # Load
    mesh = xr.open_dataset(path['mask'])
    volume = mesh['volume']
    mask = xr.open_dataset(path['mask'])['tmask'][0]
    #
    depths = mesh['gdept_1d'][0] 
    depth_condition = (depths >= 50) & (depths <= 150)
    # 
    depth_mask_3d = depth_condition.values[:, np.newaxis, np.newaxis]
    #    
    x = volume['x']  
    y = volume['y'] 
    #
    xx, yy = np.meshgrid(x, y)
    #
    # 2D Polygon Mask
    polygon_mask_2d = np.array([
        [polygon.contains(Point(xx[j, i], yy[j, i])) for i in range(len(x))]
        for j in range(len(y))
    ])

    # to 3D
    nz = volume.sizes['z']
    polygon_mask_3d = np.repeat(polygon_mask_2d[np.newaxis, :, :], nz, axis=0)

    # inside polygon  + in water + > 75 m
    combined_mask = (polygon_mask_3d & (mask.values == 1) & depth_mask_3d)

    # Applying mask
    mask_da = xr.DataArray(combined_mask, dims=volume.dims, coords=volume.coords)
    volume_in_polygon_water = volume.where(mask_da).sum().item()
    
    return volume_in_polygon_water
```

# We calculate the Colloidal particle counts

From the Parcels output file, we calculate the particles count by counting the longitudes and latitude particle points that are inside the polygon in a colloidal state and below 50 m depth.

```python
def polygon_definition_data_colloidal(ds, region_polygon,  depth_min=50):
#
    lons = ds['lon'].values
    lats = ds['lat'].values
    depths = ds['z'].values
    statuses = ds['status'].values
    #
    points_flat = np.column_stack((lons.ravel(), lats.ravel()))
    poly_path = Path(region_polygon)
    # points inside polygon
    inside_mask_flat = poly_path.contains_points(points_flat)
    inside_mask = inside_mask_flat.reshape(lons.shape)
    # >= 50 mask
    deep_mask = depths >= depth_min
    #deep_mask = depths >= depth_min
    # Colloidal status 
    status_mask = np.isin(statuses, 2) 
    #
    valid_mask = inside_mask & deep_mask & status_mask
    #
    counts = np.sum(valid_mask, axis=0) 
    #
    times = ds['time'].values
    time_index = np.nanmax(times, axis=0) 
    #
    return pd.DataFrame({'count': counts}, index=time_index)
```

Finally, concentrations are calculated by:

$[Region]$ = $\sum \frac{Particle Count}{Volume} \cdot \frac{1}{T}$

```python
def curve(t, goal, decay, to):
    return goal * (1. - np.exp(-(t+to)/decay))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

def plot_parameter_sweep(df, p):
    plt.rcParams.update({'font.size': 18})
    
    label_map = {
        'Ads': r'$Q^{w}_{MC}$',
        'Tau': r'$\tau_{c}$',
        'MC_ratio': r'$Q^{w}_{MC} / K^{w}_{MC}$',
        'Vel_C': r'($V_{SP}, V_{MP}$)$\times \nu$'
    }
    units_map = {
        'Ads' : r'Days$^{-1}$',
        'Tau' : 'Pa',
        'MC_ratio' : '',
        'Vel_C' : ''
    }

    other_params = [col for col in ['Ads', 'Tau', 'MC_ratio', 'Vel_C'] if col != p]
    vel_order = ['Lx10', 'Lx2', 'N', 'Hx1.2', 'Hx1.5', 'Hx2']

    fig, ax = plt.subplots(1, 2, figsize=(18, 7), sharex=False, gridspec_kw={'width_ratios': [2, 2]})
    grouped = df.groupby(other_params)

    # Extract the colors from the tab20 colormap
    tab20_colors = plt.cm.tab20.colors

    for i, (group_vals, group_df) in enumerate(grouped):
        # Select the color for this specific group
        group_color = tab20_colors[i % len(tab20_colors)]
        
        group_df = group_df.copy()

        if p == 'Vel_C':
            group_df[p] = pd.Categorical(group_df[p],
                                         categories=vel_order,
                                         ordered=True)
            group_df = group_df.sort_values(by=p)
            x = group_df[p].cat.codes.values
        else:
            group_df = group_df.sort_values(by=p)
            x = group_df[p].values

        y1 = group_df['[S1]'].values
        y2 = group_df['[S1]/[N1]'].values

        if len(x) < 2:
            continue
        
        params_names = [label_map[col] for col in other_params]
        params_units = [units_map[col] for col in other_params] 
        
        label = ', '.join([rf'{k}=$\mathbf{{{v}}}$ {u}' for k, v, u in zip(params_names, group_vals, params_units)])

        # Apply the specific group_color here
        ax[0].plot(x, y1*max_gpp_BDE_99, marker='o', linewidth=2, color=group_color)
        ax[1].plot(x, y2, marker='o', linewidth=2, label=label, color=group_color)

    if p == 'Vel_C':
        for i in [0, 1]: 
            ax[i].set_xticks(range(len(vel_order)))
            ax[i].set_xticklabels(vel_order)

    ax[0].set_xlabel(label_map[p]) #
    ax[0].set_ylabel(r'[S1] $g \cdot m^{-3}$')

    ax[1].set_xlabel(label_map[p])
    ax[1].set_ylabel('[S1] / [N1]')

    ax[1].legend(loc='upper right', fontsize=13, framealpha=.5)
    for axis in [ax[0], ax[1]]:
        axis.grid(True, linestyle='--', alpha=0.7)

    plt.tight_layout()
    return fig, ax
##
###
##
# PLOTS
#### REFERENCE 2007 ####
c_ref = 1.78295689e-08 
s1_n1_ref = 1.56797739

##### TAU #####
tau_fig, tau_ax = plot_parameter_sweep(df, 'Tau')
#
tau_ax[0].hlines(y=min_range_S1*min_gpp_BDE_99, xmin = np.min(all_taus), xmax = np.max(all_taus), 
                 color = 'k', linewidth = 3, linestyle = '--', alpha = .5)
tau_ax[1].axhspan(35, 56, color = 'k', alpha = .2)
# Fixed: Directly set labels on ax[0] and ax[1]
tau_ax[0].set_xlabel(r'Critical Bed Shear Stress $\tau_c$ [Pa]')
tau_ax[1].set_xlabel(r'Critical Bed Shear Stress $\tau_c$ [Pa]')

tau_ax[0].scatter(all_taus, df['[S1]']*max_gpp_BDE_99, facecolor='w', edgecolor='k')
tau_ax[1].scatter(all_taus, df['[S1]/[N1]'], facecolor='w', edgecolor='k')
#
tau_ax[0].text(x=0.0175, y=min_range_S1*min_gpp_BDE_99 + 0.01e-9, s='Min. Expected [S1]', 
               color='k', fontsize=16, 
               verticalalignment='bottom',   
               horizontalalignment='left',   
               bbox=dict(facecolor='white', edgecolor='none', alpha=0.1, pad=2))
#
tau_ax[1].text(x=0.0175, y=57, s='Expected [S1]/[N1]', 
               color='k', fontsize=16, 
               verticalalignment='bottom',   
               horizontalalignment='left',   
               bbox=dict(facecolor='white', edgecolor='none', alpha=0.1, pad=2))
#
tau_ax[1].legend(
    loc='lower center',
    bbox_to_anchor=(-0.1, 1.05),
    ncol=2,
    fontsize = 14
)

##### ADS #####
ads_fig, ads_ax = plot_parameter_sweep(df, 'Ads')
#
ads_ax[0].hlines(y=min_range_S1*min_gpp_BDE_99, xmin = np.min(all_ads), xmax = np.max(all_ads), 
                 color = 'k', linewidth = 3, linestyle = '--', alpha = .5)
ads_ax[1].axhspan(35, 56, color = 'k', alpha = .2)
# Fixed: Directly set labels
ads_ax[0].set_xlabel(r'Adsorption Rate to Marine Particles $Q^{w}_{MC}$ [$Days^{-1}$]')
ads_ax[1].set_xlabel(r'Adsorption Rate to Marine Particles $Q^{w}_{MC}$ [$Days^{-1}$]')

ads_ax[0].scatter(all_ads, df['[S1]']*max_gpp_BDE_99, facecolor='w', edgecolor='k')
ads_ax[1].scatter(all_ads, df['[S1]/[N1]'], facecolor='w', edgecolor='k')
#
ads_ax[0].text(x=1.39, y=min_range_S1*min_gpp_BDE_99 + 0.01e-9, s='Min. Expected [S1]', 
               color='k', fontsize=16, 
               verticalalignment='bottom',   
               horizontalalignment='left',   
               bbox=dict(facecolor='white', edgecolor='none', alpha=0.1, pad=2))
#
ads_ax[1].text(x=1.39, y=57, s='Expected [S1]/[N1]', 
               color='k', fontsize=16, 
               verticalalignment='bottom',   
               horizontalalignment='left',   
               bbox=dict(facecolor='white', edgecolor='none', alpha=0.1, pad=2))
#
ads_ax[1].legend(
    loc='lower center',
    bbox_to_anchor=(-0.1, 1.05),
    ncol=3,
    fontsize = 14
)
##### MC #####
mc_fig, mc_ax = plot_parameter_sweep(df, 'MC_ratio')
#
mc_ax[0].hlines(y=min_range_S1*min_gpp_BDE_99, xmin = np.min(all_mc), xmax = np.max(all_mc), 
                 color = 'k', linewidth = 3, linestyle = '--', alpha = .5)
mc_ax[1].axhspan(35, 56, color = 'k', alpha = .2)
# Fixed: Directly set labels
mc_ax[0].set_xlabel(r'Ratio Adsorption to Desorption $Q^{w}_{MC} / K^{w}_{MC}$')
mc_ax[1].set_xlabel(r'Ratio Adsorption to Desorption $Q^{w}_{MC} / K^{w}_{MC}$')

mc_ax[0].scatter(all_mc, df['[S1]']*max_gpp_BDE_99, facecolor='w', edgecolor='k')
mc_ax[1].scatter(all_mc, df['[S1]/[N1]'], facecolor='w', edgecolor='k')
mc_ax[1].legend(loc = 'upper left', fontsize=13, framealpha = .5)
mc_ax[1].legend(
    loc='lower center',
    bbox_to_anchor=(-0.1, 1.05),
    ncol=2,
    fontsize = 14
)
#
mc_ax[0].text(x=0.01, y=min_range_S1*min_gpp_BDE_99 + 0.01e-9, s='Min. Expected [S1]', 
               color='k', fontsize=16, 
               verticalalignment='bottom',   
               horizontalalignment='left',   
               bbox=dict(facecolor='white', edgecolor='none', alpha=0.1, pad=2))
#
mc_ax[1].text(x=0.01, y=57, s='Expected [S1]/[N1]', 
               color='k', fontsize=16, 
               verticalalignment='bottom',   
               horizontalalignment='left',   
               bbox=dict(facecolor='white', edgecolor='none', alpha=0.1, pad=2))
#
##### VEL #####
vel_fig, vel_ax = plot_parameter_sweep(df, 'Vel_C')
# Fixed: Directly set labels
vel_ax[0].set_xlabel(r'Sinking Velocity Change $(V_{MP}, V_{SP}) \times \nu$')
vel_ax[1].set_xlabel(r'Sinking Velocity Change $(V_{MP}, V_{SP}) \times \nu$')

# right order
mapping = {val: i for i, val in enumerate(['Lx10', 'Lx2', 'N', 'Hx1.2', 'Hx1.5', 'Hx2'])}
x_scatter = [mapping[v] for v in all_vel]

vel_ax[0].scatter(x_scatter, df['[S1]']*max_gpp_BDE_99, facecolor='w', edgecolor='k', zorder=1)
vel_ax[1].scatter(x_scatter, df['[S1]/[N1]'], facecolor='w', edgecolor='k', zorder=1)
vel_ax[1].legend(loc = 'upper left', fontsize=13, framealpha = .5)
vel_ax[1].legend(
    loc='lower center',
    bbox_to_anchor=(-0.1, 1.05),
    ncol=3,
    fontsize = 14
)
vel_ax[0].hlines(y=min_range_S1*min_gpp_BDE_99, xmin = np.min(x_scatter), xmax = np.max(x_scatter), 
                 color = 'k', linewidth = 3, linestyle = '--', alpha = .5)
vel_ax[1].axhspan(35, 56, color = 'k', alpha = .2)
#
vel_ax[0].text(x=0.1, y=min_range_S1*min_gpp_BDE_99 + 0.01e-9, s='Min. Expected [S1]', 
               color='k', fontsize=16, 
               verticalalignment='bottom',   
               horizontalalignment='left',   
               bbox=dict(facecolor='white', edgecolor='none', alpha=0.1, pad=2))
#
vel_ax[1].text(x=0.1, y=57, s='Expected [S1]/[N1]', 
               color='k', fontsize=16, 
               verticalalignment='bottom',   
               horizontalalignment='left',   
               bbox=dict(facecolor='white', edgecolor='none', alpha=0.1, pad=2))
```

```python
import numpy as np
import pandas as pd

def proportions_optimized(data):
    # 1. Extract raw arrays ONCE outside the loop (Massive speedup)
    status_vals = data.status.values
    lat_vals = data.lat.values
    lon_vals = data.lon.values
    
    # Assuming the second dimension is 'obs' based on your [:, i] slicing
    n_obs = status_vals.shape[1] 
    
    # 2. Pre-allocate a single 2D array for all results to avoid list appending
    # Shape: (n_obs, 10 columns)
    results = np.zeros((n_obs, 10), dtype=float)
    
    # 3. Fast iteration
    for i in range(n_obs):
        # Extract 1D slices for the current time step
        s_i = status_vals[:, i]
        la_i = lat_vals[:, i]
        lo_i = lon_vals[:, i]
        
        # Create valid mask
        valid_mask = np.isfinite(s_i) & np.isfinite(la_i) & np.isfinite(lo_i)
        tot = valid_mask.sum()
        
        if tot == 0:
            results[i, :-1] = np.nan  # Set all counts to NaN
            results[i, -1] = 0        # Set total to 0
            continue
            
        # Apply mask to arrays
        s_i = s_i[valid_mask]
        la_i = la_i[valid_mask]
        lo_i = lo_i[valid_mask]
        
        # Populate results directly into our pre-allocated array
        results[i, 0] = np.count_nonzero(s_i < 0)               # initial
        results[i, 1] = np.count_nonzero(s_i == 2)              # colloidal_water
        results[i, 2] = np.count_nonzero(s_i == 3)              # marine_water
        results[i, 3] = np.count_nonzero(s_i == 1)              # sewage_water
        results[i, 4] = np.count_nonzero(s_i == 12)             # colloidal_sediment
        results[i, 5] = np.count_nonzero(s_i == 13)             # marine_sediment
        results[i, 6] = np.count_nonzero(s_i == 11)             # sewage_sediment
        results[i, 7] = np.count_nonzero(s_i == 7)              # out_jdf
        results[i, 8] = np.count_nonzero(s_i == 8)              # out_js
        results[i, 9] = tot                                     # total_len

    # 4. Convert directly to DataFrame
    columns = [
        'initial', 'colloidal_water', 'marine_water', 'sewage_water',
        'colloidal_sediment', 'marine_sediment', 'sewage_sediment',
        'out_jdf', 'out_js', 'total_len'
    ]
    
    return pd.DataFrame(results, columns=columns)
```

```python
def get_system_timeseries_optimized(data, status_var='status', 
                                    target_status=2, chunk_size=500):
    """
    Computes a particle timeseries filtered by status across the entire system.
    Optimized for minimal memory usage and high execution speed via chunking.
    """
    #with xr.open_dataset(filename) as data:
    n_obs = data.sizes['obs'] 
    counts_per_time = np.zeros(n_obs, dtype=int)
    
    for start in range(0, n_obs, chunk_size):
        end = min(start + chunk_size, n_obs)
        
        status_chunk = data[status_var].isel(obs=slice(start, end)).values
        
        if isinstance(target_status, (list, tuple, np.ndarray)):
            status_mask = np.isin(status_chunk, target_status)
        else:
            status_mask = (status_chunk == target_status)
        
        if not np.any(status_mask):
            continue
            

        chunk_time_indices = np.where(status_mask)[1]
        
        absolute_time_indices = chunk_time_indices + start
        
        counts_per_time += np.bincount(absolute_time_indices, minlength=n_obs)
                
    return pd.DataFrame({'Total_Count': counts_per_time})
```
