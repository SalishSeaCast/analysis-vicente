```python
filename_AUG = '/home/vicentev/scratch/vicentev/Simulations_Runs/RESTART_Runs/R_2011/Simulation_2011_AUG.zarr'
data_zarr = zarr.open(filename_AUG, mode='r')
```

```python
# 1. Open Zarr in read-only mode
z = zarr.open(filename_AUG, mode='r')

# 2. Determine the 'Safe Limit' (3702)
safe_limit = z['H_vel_2'].shape[1] 
n_trajectories = z['trajectory'].shape[0] # Should be 2976

print(f"Dimensions -> Trajectories: {n_trajectories}, Obs: {safe_limit}")

# 3. --- FIX: GENERATE CORRECT TIME MANUALLY ---
# We create the time array from scratch based on your specifications.
# Start: 2008-12-30T22:30:00, Frequency: 2 Hours
clean_time_1d = pd.date_range(
    start='2008-12-30T20:30:00', 
    periods=safe_limit, 
    freq='2h'
)

# Broadcast this 1D time array to 2D (trajectory, obs) so it matches (2976, 3702)
clean_time_2d = np.tile(clean_time_1d, (n_trajectories, 1))
# -----------------------------------------------

# 4. List of variables to load (excluding 'time' since we made it manually)
var_names = [
    'bat_particle', 'crit', 'e3t', 'fact', 'fmask', 'H_vel_2', 
    'lat', 'lon', 'release_time', 'status', 'stuck', 
    'tmask', 'totaldepth', 'umask', 'uvalue', 'vmask', 'vvalue', 'wvalue', 'z'
]

data_dict = {}

# 5. Load other variables normally
for var in var_names:
    arr = z[var]
    if len(arr.shape) == 2:
        # Slice to safe limit
        data_dict[var] = (("trajectory", "obs"), arr[:, :safe_limit])
    elif len(arr.shape) == 1:
        data_dict[var] = (("trajectory",), arr[:])

# 6. Add our manually created time variable
data_dict['time'] = (("trajectory", "obs"), clean_time_2d)

# 7. Create the Dataset
data_AUG = xr.Dataset(
    data_vars=data_dict,
    coords={
        "trajectory": np.arange(n_trajectories),
        "obs": np.arange(safe_limit)
    }
)

print("Dataset created with correct datetime64[ns] time!")
```
